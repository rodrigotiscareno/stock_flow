```python
tickerSymbol = 'AAPL'
tickerData = yf.Ticker(tickerSymbol)
```

### Granular, Interval Historical Data

```python
tickerData.history(period='1d', interval='15m')

# [1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo, 3mo]
```

```python
dividends = tickerData.dividends

pd.DataFrame(dividends).reset_index()
```

```python
splits = tickerData.splits

pd.DataFrame(splits).reset_index()
```
